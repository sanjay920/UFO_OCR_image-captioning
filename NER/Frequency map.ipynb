```python
import pandas as pd
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df = pd.read_csv('PA2_UFO_sightings.csv')
```

```python
text_list = df.text.tolist()
```

```python
text_list = [str(t) for t in text_list]
```

```python
len(text_list)
```

```python
text_list[1]
```

```python
word_tokenize(text_list[1])
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
word_list = []
for lst in text_list:
#     tokenized = word_tokenize(lst)
    tokenized = tokenizer.tokenize(lst)
    to_extend = [x for x in tokenized if x not in stop_words]
    word_list.extend(to_extend)
    
```

```python
from collections import Counter
```

```python
frequency_map = Counter(word_list)
```

```python
for i,j in frequency_map.most_common():
    print j,i
```
