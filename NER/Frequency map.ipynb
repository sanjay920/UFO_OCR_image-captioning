```python
import pandas as pd
from nltk import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import numpy as np
import enchant
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df = pd.read_csv('PA2_UFO_sightings.csv')
```

```python
text_list = df.text.tolist()
```

```python
text_list = [str(t) for t in text_list]
```

```python
len(text_list)
```

```python
text_list[1]
```

```python
word_tokenize(text_list[1])
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
word_list = []
for lst in text_list:
#     tokenized = word_tokenize(lst)
    tokenized = tokenizer.tokenize(lst)
    to_extend = [x for x in tokenized if x not in stop_words]
    word_list.extend(to_extend)
    
```

```python
from collections import Counter
```

```python
frequency_map = Counter(word_list)
```

```python
ignore = []
d = enchant.Dict("en_US")
for i,j in frequency_map.items():
    if len(i) < 3 or not d.check(i):
        ignore.append(i)
```

```python
ignore
```

```python
for i in ignore:
    del frequency_map[i]
```

```python
for i,j in frequency_map.most_common(100):

    print j,i
```

```python
labels, values = zip(*frequency_map.most_common(50))

indexes = np.arange(len(labels))
width = 1

plt.bar(indexes, values, width)
plt.xticks(indexes + width * 0.5, labels)
plt.xticks(rotation=90)
plt.savefig('word_frequency_hist.eps', format='eps', dpi=1000, bbox_inches='tight')
plt.show()
```
